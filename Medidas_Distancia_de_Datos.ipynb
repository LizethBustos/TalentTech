```python
# Supongamos que tenemos dos puntos en un plano cartesiano, A(2, 3) y B(5, 7).
# Queremos calcular la distancia euclidiana entre estos dos puntos.

def euclidean_distance(x1, x2, y1, y2):
    return ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5

# Ejemplo de uso
x1, y1 = 2, 3
x2, y2 = 5, 7
print("Distancia Euclidiana entre A y B:", euclidean_distance(x1, x2, y1, y2))

 #zip para iterar sobre las coordenadas de ambos puntos simult√°neamente.
```

```python
# Consideremos dos ubicaciones en una cuadr√≠cula de la ciudad, donde A(1, 4) representa una ubicaci√≥n y B(7, 9) otra.
# Queremos calcular la distancia de Manhattan entre estas dos ubicaciones.

def manhattan_distance(x1, x2, y1, y2):
    return abs(x2 - x1) + abs(y2 - y1)

# Ejemplo de uso
x1, y1 = 1, 4
x2, y2 = 7, 9
print("Distancia de Manhattan entre A y B:", manhattan_distance(x1, x2, y1, y2))
```

**Distancia de Hamming**

image.png

```python
# Supongamos que estamos trabajando con datos categ√≥ricos y tenemos dos secuencias binarias A y B.
# Queremos calcular la distancia de Hamming entre estas dos secuencias.
def hamming_distance(sequence1, sequence2):
    if len(sequence1) != len(sequence2):
        raise ValueError("Las secuencias deben tener la misma longitud")
    distance = 0
    for i in range(len(sequence1)):
        if sequence1[i] != sequence2[i]:
            distance += 1
    return distance

# Ejemplo de uso
sequence_a = "1011101"
sequence_b = "1001001"
print("Distancia de Hamming entre A y B:", hamming_distance(sequence_a, sequence_b))

#Si es diferente de 0 las secuencias son diferentes
```


**Distancia de Chebyshev**

image.png

```python
# Supongamos que estamos trabajando con datos espaciales y tenemos dos puntos en un plano cartesiano.
# Queremos calcular la distancia de Chebyshev entre estos dos puntos.

def chebyshev_distance(x1, y1, x2, y2):
    return max(abs(x1 - x2), abs(y1 - y2))

# Ejemplo de uso
x1, y1 = 1, 3
x2, y2 = 5, 7
print("Distancia de Chebyshev entre los puntos:", chebyshev_distance(x1, y1, x2, y2))
```

**Distancia de Canberra**

image.png

```python
def canberra_distance(vector1, vector2):
    if len(vector1) != len(vector2):
        raise ValueError("Los vectores deben tener la misma longitud")
    distance = 0
    for i in range(len(vector1)):
        distance += abs(vector1[i] - vector2[i]) / (abs(vector1[i]) + abs(vector2[i]))
    return distance

# Ejemplo de uso
vector1 = [1, 2, 3]
vector2 = [4, 5, 6]
print("Distancia de Canberra entre los vectores:", canberra_distance(vector1, vector2))
```

**Distancia Coseno**

image.png

```python
def cosine_similarity(vector1, vector2):
    dot_product = sum(x * y for x, y in zip(vector1, vector2))
    magnitude1 = sum(x ** 2 for x in vector1) ** 0.5
    magnitude2 = sum(y ** 2 for y in vector2) ** 0.5
    return dot_product / (magnitude1 * magnitude2)

# Ejemplo de uso
vector1 = [3, 4, 2]
vector2 = [1, 2, 5]
print("Similitud del coseno entre los vectores:", cosine_similarity(vector1, vector2))
```

**Distancia de Jaccard**

image.png

```python
def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union

# Ejemplo de uso
set1 = {1, 2, 3}
set2 = {2, 3, 4}
print("Similitud de Jaccard entre los conjuntos:", jaccard_similarity(set1, set2))
```
