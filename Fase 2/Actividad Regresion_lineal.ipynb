```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creamos el DataFrame
data = pd.DataFrame({'X': [1, 2, 2, 3, 4, 4, 5, 6],
                     'Y': [2, 3, 4, 4, 4, 6, 5, 7]})

# Calculamos la media de X y de Y
mean_x = np.mean(data['X'])
mean_y = np.mean(data['Y'])

# Calculamos los trminos necesarios para la frmula de regresin lineal
data['Covarianza'] = (data['X'] - mean_x) * (data['Y'] - mean_y)
data['Varianza_X'] = (data['X'] - mean_x) ** 2

# Calculamos los coeficientes de la regresin lineal
m = data['Covarianza'].sum() / data['Varianza_X'].sum()
b = mean_y - m * mean_x

# Aplicamos la frmula de regresin lineal para predecir los valores de Y
data['Prediccion_Y'] = m * data['X'] + b

# Graficamos los datos y la lnea de regresin
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color='blue', label='Datos reales') #puntos
plt.plot(data['X'], data['Prediccion_Y'], color='red', label='Lnea de regresin')
plt.title('Regresin Lineal: Y en funcin de X')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Mostramos los coeficientes de la regresin lineal
print(data)
print(f"Coeficiente m: {m}")
print(f"Coeficiente b: {b}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creamos el DataFrame
data = pd.DataFrame({'X': [1, 2, 2, 3, 4, 4, 5, 6],
                     'Y': [2, 3, 4, 4, 4, 6, 5, 7]})

n = len(data['X'])

data['X*Y'] = data['X'] * data['Y']
data['X^2'] = data['X'] ** 2

#suma de todos los valores de cada columna del DataFrame
sum_X = data['X'].sum()
sum_Y = data['Y'].sum()
sum_XY = data['X*Y'].sum()
sum_XX = data['X^2'].sum()

#sacamos los promedios
prom_X = data['X'].mean()
prom_Y = data['Y'].mean()

num = n * sum_XY - sum_X * sum_Y
den = n * sum_XX - (sum_X)**2

m = num / den
b = prom_Y - m * prom_X

x_valores = np.linspace(min(data['X']),max(data['X']))
y_valores = m * x_valores + b

# Graficamos los datos y la lnea de regresin
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color='blue', label='Datos reales') #puntos
plt.plot(x_valores, y_valores, color='red', label='Lnea de regresin')
plt.title('Regresin Lineal: Y en funcin de X')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Mostramos los coeficientes de la regresin lineal
print(data)
print(num)
print(den)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creamos el DataFrame
data = pd.DataFrame({'X': [1, 2, 2, 3, 4, 4, 5, 6],
                     'Y': [2, 3, 4, 4, 4, 6, 5, 7]})

n = len(data['X'])

data['X*Y'] = data['X'] * data['Y']
data['X^2'] = data['X'] ** 2

#Se inserta el nuevo registro
fila_suma_total = pd.DataFrame({'X': [data['X'].iloc[-1]], 'Y': [data['Y'].iloc[-1]], 'X*Y' : [data['X*Y'].iloc[-1]], 'X^2' : [data['X^2'].iloc[-1]]})
data = pd.concat([data, fila_suma_total], ignore_index=True)

#sacamos los promedios
prom_X = data['X'].mean()
prom_Y = data['Y'].mean()

#utilizamos la nueva fila
num = n * data['X*Y'].iloc[-1] - data['X'].iloc[-1] * data['Y'].iloc[-1]
den = n * data['X^2'].iloc[-1] - (data['X'].iloc[-1])**2

m = num / den
b = prom_Y - m * prom_X

x_valores = np.linspace(min(data['X']),max(data['X']))
y_valores = m * x_valores + b

# Graficamos los datos y la lnea de regresin
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color='blue', label='Datos reales') #puntos
plt.plot(x_valores, y_valores, color='red', label='Lnea de regresin')
plt.title('Regresin Lineal: Y en funcin de X')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Mostramos los coeficientes de la regresin lineal
print(data)
print("Pendiente (m):", m)
print("Interseccin (b):", b)
```
