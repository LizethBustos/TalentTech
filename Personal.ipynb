```python
import pandas as pd

# Datos de ejemplo
datos = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril'],
    'Ventas': [15000, 20000, 18000, 22000]
}

# Crear DataFrame
df_ventas = pd.DataFrame(datos)
df_ventas.head(5)
```

```python
# Calcular el total de ventas por mes
ventas_por_mes = df_ventas.groupby('Mes')['Ventas'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

# Gráfico de barras de ventas por mes
plt.figure(figsize=(10, 6))
plt.bar(ventas_por_mes['Mes'], ventas_por_mes['Ventas'], color='skyblue')
plt.title('Ventas por Mes')
plt.xlabel('Mes')
plt.ylabel('Ventas ($)')
plt.show()
```
